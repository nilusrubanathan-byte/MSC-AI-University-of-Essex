```python
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt

test = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
# Check for missing values in the test set
missing_test = test.isnull().sum()

# Get the percentage of missing values
missing_test = missing_test[missing_test > 0]
missing_test_percentage = (missing_test / len(test)) * 100

print("Missing values in the test set:")
print(missing_test_percentage.sort_values(ascending=False))
```

```python
# Visualize missing values
msno.matrix(test)
plt.title('Missing Value Matrix for Test Data')
plt.show()

msno.bar(test)
plt.title('Missing Value Bar Chart for Test Data')
plt.show()
```
