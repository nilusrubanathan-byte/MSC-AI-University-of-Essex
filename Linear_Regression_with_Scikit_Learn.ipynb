```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Load the datasets
df_population = pd.read_csv('/content/Unit04 Global_Population.csv')
df_gdp = pd.read_csv('/content/Unit04 Global_GDP.csv')

# Replace '..' with NaN in the population dataframe and convert to numeric
df_population.replace('..', np.nan, inplace=True)
years = [str(year) for year in range(2001, 2021)]
for year in years:
    df_population[year] = pd.to_numeric(df_population[year], errors='coerce')

# Handle missing values by filling with the mean of numeric columns only
df_population.fillna(df_population.mean(numeric_only=True), inplace=True)
df_gdp.fillna(df_gdp.mean(numeric_only=True), inplace=True)

# Select years 2001-2020 and calculate the mean population
df_population_years = df_population[['Country Name'] + years]
mean_population = df_population_years.set_index('Country Name')[years].mean(axis=1, numeric_only=True).reset_index(name='Mean Population (2001-2020)')

# Select years 2001-2020 and calculate the mean GDP
df_gdp_years = df_gdp[['Country Name'] + years]
mean_gdp = df_gdp_years.set_index('Country Name')[years].mean(axis=1, numeric_only=True).reset_index(name='Mean GDP (2001-2020)')

# Merge the mean population and mean GDP data
merged_df = pd.merge(mean_population, mean_gdp, on='Country Name')

# Calculate mean per capita GDP (Mean GDP / Mean Population)
merged_df['Mean Per Capita GDP (2001-2020)'] = merged_df['Mean GDP (2001-2020)'] / merged_df['Mean Population (2001-2020)']

# Generate a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Mean Population (2001-2020)'], merged_df['Mean Per Capita GDP (2001-2020)'])

# Label the axes and add a title
plt.xlabel('Mean Population (2001-2020)')
plt.ylabel('Mean Per Capita GDP (2001-2020)')
plt.title('Mean Population vs. Mean Per Capita GDP (2001-2020)')

# Calculate the linear regression line
independent_variable = merged_df['Mean Population (2001-2020)']
dependent_variable = merged_df['Mean Per Capita GDP (2001-2020)']
slope, intercept, rvalue, pvalue, stderr = linregress(independent_variable, dependent_variable)

# Add the regression line to the scatter plot
x = np.array(independent_variable)
plt.plot(x, intercept + slope * x, color='red', label=f'Regression Line (R-value: {rvalue:.2f})')
plt.legend()

# Display the scatter plot with the regression line
plt.show()

# Print the correlation coefficient
print(f"Pearson Correlation Coefficient: {rvalue}")
```

## Analyze correlation

### Subtask:
Investigate the correlation between the mean population and mean per capita GDP using visualization and calculate the Pearson Correlation Coefficient.


**Reasoning**:
Generate a scatter plot to visualize the relationship between mean population and mean per capita GDP, and then calculate the Pearson correlation coefficient.



```python
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np

# Generate a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Mean Population (2001-2020)'], merged_df['Mean Per Capita GDP (2001-2020)'])

# Label the axes and add a title
plt.xlabel('Mean Population (2001-2020)')
plt.ylabel('Mean Per Capita GDP (2001-2020)')
plt.title('Mean Population vs. Mean Per Capita GDP (2001-2020)')

# Calculate the linear regression line
independent_variable = merged_df['Mean Population (2001-2020)']
dependent_variable = merged_df['Mean Per Capita GDP (2001-2020)']
slope, intercept, rvalue, pvalue, stderr = linregress(independent_variable, dependent_variable)

# Add the regression line to the scatter plot
x = np.array(independent_variable)
plt.plot(x, intercept + slope * x, color='red', label=f'Regression Line (R-value: {rvalue:.2f})')
plt.legend()

# Display the scatter plot with the regression line
plt.show()

# Calculate the Pearson correlation coefficient (already calculated in linregress)
correlation = rvalue

# Print the correlation coefficient
print(f"Pearson Correlation Coefficient: {correlation}")
```

**Reasoning**:

```python
from scipy.stats import linregress

# Extract the independent and dependent variables
independent_variable = merged_df['Mean Population (2001-2020)']
dependent_variable = merged_df['Mean Per Capita GDP (2001-2020)']

# Perform linear regression
slope, intercept, rvalue, pvalue, stderr = linregress(independent_variable, dependent_variable)

# Print the results
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-value: {rvalue}")
print(f"P-value: {pvalue}")
print(f"Standard error: {stderr}")
```
