## Load the data

### Subtask:
Load the `iris.csv` dataset into a pandas DataFrame.


**Reasoning**:
Load the data into a pandas DataFrame and display its head and info.



```python
import pandas as pd

df = pd.read_csv('/content/Unit06 iris.csv')
display(df.head())
display(df.info())
```

## Prepare the data

### Subtask:
Inspect the data, handle any missing values, and remove non-numeric columns that are not suitable for K-Means clustering.


**Reasoning**:
Inspect the data for missing values and remove the non-numeric 'species' column, then display the resulting DataFrame's head and info.



```python
display(df.isnull().sum())
df_numeric = df.drop('species', axis=1)
display(df_numeric.head())
display(df_numeric.info())
```

## Determine the optimal number of clusters (k)

### Subtask:
Use methods like the elbow method or silhouette score to help determine a reasonable value for K.


**Reasoning**:
Calculate and plot the WCSS and silhouette scores for a range of K values to determine a reasonable K for clustering.



```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

wcss = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_numeric)
    wcss.append(kmeans.inertia_)
    score = silhouette_score(df_numeric, kmeans.labels_)
    silhouette_scores.append(score)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(k_range, wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')

plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')

plt.tight_layout()
plt.show()
```

## Perform k-means clustering

### Subtask:
Apply the K-Means algorithm with different values of K, including K=3 as requested.


**Reasoning**:
Apply the KMeans algorithm with n_clusters=3 and n_clusters=5 as requested and store the resulting labels in the original dataframe.



```python
from sklearn.cluster import KMeans

kmeans_k3 = KMeans(n_clusters=3, random_state=42, n_init='auto')
kmeans_k3.fit(df_numeric)
df['kmeans_labels_k3'] = kmeans_k3.labels_

kmeans_k5 = KMeans(n_clusters=5, random_state=42, n_init='auto')
kmeans_k5.fit(df_numeric)
df['kmeans_labels_k5'] = kmeans_k5.labels_

display(df.head())
```

## Evaluate and visualize the results

Evaluate the clustering performance, especially for K=3, by comparing the clusters to the original species labels. Visualize the clusters if possible (e.g., using PCA for dimensionality reduction if needed).


**Reasoning**:
Calculate a confusion matrix and evaluation metrics for K=3 clustering, perform PCA for dimensionality reduction, and create scatter plots to visualize the clustering results and compare them with the true species labels.



```python
from sklearn.metrics import confusion_matrix, homogeneity_score, completeness_score, v_measure_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 1. Calculate confusion matrix or cross-tabulation for K=3
conf_matrix_k3 = confusion_matrix(df['species'], df['kmeans_labels_k3'])
print("Confusion Matrix for K=3 Clustering vs. Original Species:")
print(conf_matrix_k3)

# Calculate cross-tabulation as an alternative visualization
cross_tab_k3 = pd.crosstab(df['species'], df['kmeans_labels_k3'])
print("\nCross-Tabulation for K=3 Clustering vs. Original Species:")
print(cross_tab_k3)

# 2. Calculate evaluation metrics for K=3 clustering
homogeneity_k3 = homogeneity_score(df['species'], df['kmeans_labels_k3'])
completeness_k3 = completeness_score(df['species'], df['kmeans_labels_k3'])
v_measure_k3 = v_measure_score(df['species'], df['kmeans_labels_k3'])

print(f"\nHomogeneity Score for K=3: {homogeneity_k3:.4f}")
print(f"Completeness Score for K=3: {completeness_k3:.4f}")
print(f"V-Measure Score for K=3: {v_measure_k3:.4f}")

# 3. Perform PCA to reduce dimensionality to 2 components
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_numeric)
df_pca = pd.DataFrame(data=principal_components, columns=['principal_component_1', 'principal_component_2'])

# Add the cluster labels and original species to the PCA DataFrame
df_pca['kmeans_labels_k3'] = df['kmeans_labels_k3']
df_pca['species'] = df['species']

# 4. Create a scatter plot colored by kmeans_labels_k3
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
scatter1 = plt.scatter(df_pca['principal_component_1'], df_pca['principal_component_2'],
                       c=df_pca['kmeans_labels_k3'], cmap='viridis', alpha=0.7)
plt.title('PCA of Iris Dataset - K=3 KMeans Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter1, label='KMeans Cluster')

# 5. Create a scatter plot colored by original species
plt.subplot(1, 2, 2)
scatter2 = plt.scatter(df_pca['principal_component_1'], df_pca['principal_component_2'],
                       c=df_pca['species'].astype('category').cat.codes, cmap='viridis', alpha=0.7) # Map species to numeric for colormap
plt.title('PCA of Iris Dataset - Original Species')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter2, ticks=range(len(df_pca['species'].unique())),
             ticklabels=df_pca['species'].unique(), label='Species')

plt.tight_layout()
plt.show()
```
